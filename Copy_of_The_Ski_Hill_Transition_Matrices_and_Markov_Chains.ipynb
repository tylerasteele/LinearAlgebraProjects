Modeling an Epidemic

This problem describes a disease with a known infection rate, mortality rate, and understanding of the reinfection process. With all of these knowns, at some point in time, or at the beginning of the epidemic, we can quantify how many people will be at each of the four possible states where a person can be susceptible, infected, recovered and immune, or deceased. 

The matrix is similar to the previous example where we have this idea of going from as columns and going to as rows. For example, we have people going from being susceptible to being infected or, sadly, going from being infected to deceased. Since there are four states, we will have a 4x4 matrix.

The matrix is as follows:

```
# S = Susceptible, I = Infected, R = Recovered, D = Deceased
            #    S     I    R  D 
n = np.matrix([[0.95, 0.04, 0, 0],
               [0.05, 0.85, 0, 0],
               [0, 0.1, 1, 0],
               [0, 0.01, 0, 1]])

print(n)
```

The United States has an approximate population of 330 million people. Our initial state will be 330 million susceptible people where each iteration is one day.

```
# In millions
n0 = np.matrix([[330], [0], [0], [0]])

Q = np.c_[n0, n*n0, n**2 * n0, n**3 * n0, n**4 * n0, n**5 * n0, n**10 * n0, n**15 * n0, n**30 * n0, n**60 * n0, n**100 * n0 ]
```

```
x  = [0, 1, 2, 3, 4, 5, 10, 15, 30, 60, 100]
plt.plot(x, np.array(Q[0])[0], label="Susceptible")
plt.plot(x, np.array(Q[1])[0], label="Infected")
plt.plot(x, np.array(Q[2])[0], label="Recovered")
plt.plot(x, np.array(Q[3])[0], label="Deceased")
plt.plot()

plt.xlabel("x axis")  # Update this label
plt.ylabel("y axis")  # Update this label
plt.title("Line Graph Example")  # Update the title
plt.legend()
plt.show()
```

As we see above, the disease spreads very quickly early on. In the first 10 days, we see rapid infection of approximately 75 million Americans. After this peak, it slowly trails off as people either become immune or die. Interestingly, the death rate increases very slowly and almost in a constant manner for the first 100 days though it slows slightly around day 60. The recovery graph has a high rate of change in the beginning and trails of and the number of people who can become recovered decreases.

In total, the epidemic kills about 40 million people and is almost entirely eradicated in 100 days.
